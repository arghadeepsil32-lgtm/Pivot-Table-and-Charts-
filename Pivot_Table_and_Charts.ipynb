4. **Explain the difference between Sorting and Filtering with an example from this dataset.**

Answer ▶

Sorting ▶ Sorting changes the order of the rows based on the values in a specific column.

Example Scenario ▶ Imagine you want to see which sale made the most money. You would Sort Column D (Sales) from "Largest to Smallest."

What would happen ▶

- Row 16 (May 13, Sales ₹497.00) is currently near the bottom. After sorting, this row would jump to the very top because it has the highest value.

-  Row 10 (March 17, Sales ₹76.00) would drop to the very bottom because it has the lowest value.

- Result: You still see all 20 rows, just in a different order.

On the other hand,

Filtering ▶ Filtering sepates specific data by hiding rows that do not meet the criteria.

Example Scenario ▶ Imagine you are the manager for the West region and don't care about the other sales right now. You would apply a Filter on Column B (Region) to show only "West".

What would happen ▶

- Rows 3, 4, 8, 10, 12 (which are "South") would disappear from view.

- Rows 14 (East) and 20 (North) would also disappear.

 - Result: You would only see a short list containing rows 2, 5, 6, 7, 9, 11, 13, 15, 18, and 19. The other data is still there, just hidden until you clear the filter.

 Basicaaly, In case of Sorting, all rows remain visible but In case of Filtering, only rows matching the criteria are visible to the user.

7. **Why are Dashboards important for decision-making in business analysis?**

ANswer ▶

Because they convert complicated, unprocessed data into an understandable, actionable, and summarized visual summary, dashboards are crucial for business analytical decision-making. They serve as the link between enormous volumes of data and the strategic decisions that propel company success.

Dashboards successfully eliminate speculation and replace it with data-driven assurance, resulting in more intelligent, quicker, and logical strategic choices.
